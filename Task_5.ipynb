# Data visualization

```python
plt.figure(figsize=(15,10))
sns.countplot(x="season",data=df1, palette="bright")
plt.ylabel("number of matches")
plt.show()
```

**Most of the matchs played in  2011 to 2013.**

```python
mostteam1winnerruns=df1.groupby("team1").win_by_runs.sum()
mostteam1winnerwicket=df1.groupby("team1").win_by_wickets.sum()
mostteam2winnerruns=df1.groupby("team2").win_by_runs.sum()
mostteamwinnerwicket=df1.groupby("team2").win_by_wickets.sum()
```

```python
plt.figure(figsize=(12,6))
plt.ylabel("number of wins")
mostteam1winnerruns.plot(kind="bar", color="turquoise")
plt.show()
```

**Mumbai Indians,Chennai Super Kings,Royal Challengers Bangalore,Kolkata Knight Riders these teams are win by run maximum times.**

```python
plt.figure(figsize=(12,6))
plt.ylabel("number of wins")
mostteam1winnerwicket.plot(kind="bar", color="coral")
plt.show()
```

**Maximum times Kings XI Punjab,Kolkata Knight Riders,Royal Challengers Bangalore,Mumbai Indians these temas win by wickets.**

```python
plt.figure(figsize=(12,6))
plt.ylabel("number of wins")
mostteam2winnerruns.plot(kind="bar", color="indianred")
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.ylabel("number of wins")
mostteamwinnerwicket.plot(kind="bar", color="salmon")
plt.show()
```

**In team2 Kolkata Knight Riders,Mumbai Indians,Royal Challengers Bangalore,Delhi Daredevils,Chennai Super Kings,Kings XI Punjab these teams are win by run maximum times.**

```python
plt.figure(figsize=(12,6))
y=df1.winner.value_counts()
sns.barplot(x=y.index, y=y, palette="bright")
plt.xticks(rotation=45)
plt.show()
```

**This plot show maximum times winner teams.**

```python
topplayers=df1.player_of_match.value_counts()
print(topplayers.idxmax())
```

**CH Gayle is top player in all matches**

```python
plt.figure(figsize=(14,8))
topplayers=df1.player_of_match.value_counts()[:12]
sns.barplot(x = topplayers.index, y = topplayers, orient='v', palette="rocket")
plt.xticks(rotation=45)
plt.show()
```

**CH Gayle,AB deVilliers played maximum match and RG Sharma,MS Dhoni,DA Warner played same number of match.**

```python
plt.figure(figsize=(12,6))
sns.countplot(x='venue', data=df1)
plt.xticks(rotation='vertical')
plt.show()
```

**Most of the match played in Eden Gardens,M Chinnawwamy Stadium,Wankhed Stadium,Feroz Shah Kotla.**

```python
plt.figure(figsize=(12,6))
sns.countplot(x='venue', data=df1)
plt.xticks(rotation='vertical')
plt.ylabel("number of wins per team")
plt.show()
```
